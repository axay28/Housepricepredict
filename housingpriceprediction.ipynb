```python
def isfloatn(n):
    try:
        float(n)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(isfloatn)]
```

```python
def convert(x):
    st=x.split('-')
    if len(st)==2:
        return (float(st[0])+float(st[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4=df3.copy()
```

```python
df4['total_sqft']=df4['total_sqft'].apply(convert)
df4.head(4)
```

```python
df6 = df5[~(df5.total_sqft/df5.Bedrooms<300)]
df6.shape
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
        #print(key)
    print(subdf)
    print(df_out)
    return df_out
    
df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def remove_Bedroom_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        Bedrooms_stats = {}
        for Bedrooms, Bedrooms_df in location_df.groupby('Bedrooms'):
            Bedrooms_stats[Bedrooms] = {
                'mean': np.mean(Bedrooms_df.price_per_sqft),
                'std': np.std(Bedrooms_df.price_per_sqft),
                'count': Bedrooms_df.shape[0]
            }
        for Bedrooms, Bedrooms_df in location_df.groupby('Bedrooms'):
            stats = Bedrooms_stats.get(Bedrooms-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, Bedrooms_df[Bedrooms_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_Bedroom_outliers(df7)
# df8 = df7.copy()
df8.shape
```

```python
df8['bath'].unique()
```

```python
df9 = df8[df8.bath<df8.Bedrooms+2]
df9.shape
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(3)
```

```python
dummies = pd.get_dummies(df10.location)
dummies.head(3)
```

```python
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head()
```

```python
df12 = df11.drop('location',axis='columns')
df12.head(2)
```

```python
X = df12.drop(['price'],axis='columns')
X.head(3)
```

```python
X = df12.drop(['price'],axis='columns')
X.head(3)
```

```python
y = df12.price
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
